```python
import math

# -------------------------------------------------
# 1 距離ベースのスコア
# -------------------------------------------------
def length_score(text, target=100):
    length = len(text)
    diff = abs(length - target)
    score = max(0, 1 - diff / target)
    return score

# -------------------------------------------------
# 2 許容範囲 ± tolerance
# -------------------------------------------------
def evaluate_summary_length(text, target=100, tolerance=5):
    length = len(text)
    diff = abs(length - target)

    if diff <= tolerance:
        return 1.0

    score = max(0, 1 - (diff - tolerance) / target)
    return score

# -------------------------------------------------
#  テスト用コード
# -------------------------------------------------
if __name__ == "__main__":
    sample_text = "「春」をキーワードに星加アナウンサーが松山・石手を歩く。春の河川敷でビーチバレーの特訓をする若者やこの春入学を控える新一年生、自宅の壁一面にレトロ看板を「貼る」夫婦に遭遇。食欲の春！一日７５０本売れる食パンに、老舗うどん店の「メロンうどん」とは？"
    print("文字数:", len(sample_text))

    print("1 simple score:", round(length_score(sample_text), 4))
    print("2 practical score:", round(evaluate_summary_length(sample_text), 4))
```
